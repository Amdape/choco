```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("./Chocolate Sales.csv")
df.head()
```

```python
df.Date= pd.to_datetime(df.Date, format="%d-%b-%y")
```

```python
#df['Month'] = df.Date.dt.month#
df["Month"] = df["Date"].dt.strftime("%B")
```

```python
df.Amount= df.Amount.str.replace("$","").str.replace(",","").str.strip().astype("float")
```

```python
df.Country.value_counts()
```

```python
# Count of each country
country_counts = df["Country"].value_counts()

# Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(country_counts, labels=country_counts.index, autopct="%1.1f%%", startangle=140, colors=plt.cm.Pastel1.colors)
plt.title("Country Distribution", fontsize=16, fontweight="bold")
plt.axis("equal")  # Equal aspect ratio to ensure the pie is a circle
plt.show()
```

```python
# Count of each country
country_counts = df["Country"].value_counts()

# Define colors and explode settings
colors = ['#66c2a5', '#fc8d62', '#8da0cb', '#e78ac3', '#a6d854']
explode = [0.05] * len(country_counts)  # Slightly "pop" all slices

# Plot the pie chart
plt.figure(figsize=(10, 8))
wedges, texts, autotexts = plt.pie(
    country_counts,
    labels=country_counts.index,
    autopct="%1.1f%%",
    startangle=140,
    colors=colors,
    explode=explode,
    shadow=True,
    textprops={'fontsize': 12, 'fontweight': 'bold'}
)

# Customize label and percentage text
for text in texts:
    text.set_fontsize(14)
    text.set_fontweight("bold")

for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontsize(13)
    autotext.set_fontweight("bold")

# Title
plt.title("🌍 Country Distribution", fontsize=18, fontweight="bold", color="#333")

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

# Show the chart
plt.tight_layout()
plt.show()
```

```python
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', ]
               #'September', 'October', 'November', 'December']
df['Month'] = pd.Categorical(df['Month'], categories=month_order, ordered=True)

monthly_sum = df.groupby("Month")['Amount'].sum()
```

```python
monthly_sum
```

```python
# Set seaborn style
sns.set(style="whitegrid")

# Plot line chart
plt.figure(figsize=(10, 6))
sns.lineplot(x=monthly_sum.index, y=monthly_sum.values)
plt.title('Monthly Sum of Amount')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.xticks(rotation=45)
plt.show()
```

```python
# Set a custom style using seaborn
sns.set(style="darkgrid")

# Plot with additional styles
plt.figure(figsize=(10, 6))
sns.lineplot(x=monthly_sum.index, y=monthly_sum.values, 
             marker='o',         # Adding markers to the line
             color='purple',     # Line color
             linewidth=2,        # Line width
             markersize=8,       # Marker size
             markerfacecolor='red',  # Marker face color
             markeredgewidth=2)  # Marker edge width

# Title and labels with custom font sizes
plt.title('Monthly Sum of Amount', fontsize=16, fontweight='bold', color='navy')
plt.xlabel('Month', fontsize=14, fontweight='bold', color='darkred')
plt.ylabel('Amount', fontsize=14, fontweight='bold', color='darkred')

# Custom x-tick rotation and style
plt.xticks(rotation=45, fontsize=12, ha='right', color='black')

# Custom y-tick style
plt.yticks(fontsize=12, color='black')

# Add a grid with a specific line style
plt.grid(True, linestyle='--', linewidth=0.7, color='gray')

# Show the plot
plt.tight_layout()  # Automatically adjust layout
plt.show()
```

```python
# Group and sum
country_amount = df.groupby("Country")["Amount"].sum().sort_values(ascending=False)

# Define a custom format function to show both value and percentage
def autopct_format(pct, all_vals):
    absolute = int(round(pct / 100. * sum(all_vals)))
    return f"${absolute:,}\n({pct:.1f}%)"

# Styling
colors = plt.cm.Pastel2.colors
explode = [0.05] * len(country_amount)

# Plot
plt.figure(figsize=(10, 8))
wedges, texts, autotexts = plt.pie(
    country_amount,
    labels=country_amount.index,
    autopct=lambda pct: autopct_format(pct, country_amount),
    startangle=140,
    colors=colors,
    explode=explode,
    shadow=True,
    textprops={'fontsize': 12, 'fontweight': 'bold'}
)

# Improve label aesthetics
for text in texts:
    text.set_fontsize(13)
    text.set_fontweight("bold")

for autotext in autotexts:
    autotext.set_fontsize(11)
    autotext.set_color('black')
    autotext.set_fontweight("semibold")

plt.title("💵 Total Amount by Country", fontsize=18, fontweight="bold")
plt.axis("equal")
plt.tight_layout()
plt.show()
```
